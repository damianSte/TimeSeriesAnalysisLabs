```python
import numpy as np
import pandas as pd
```

```python
df =pd.read_csv('IBM.csv', index_col='Date', parse_dates = True)
```

```python
df_close = df['Close']
df_close.plot()
```

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pandas.plotting import lag_plot
```

```python
plot_acf(df_close, lags = 17, zero = False); #lags to establish periodicity
```

```python
df['dif1'] = df['Close'].diff()
df['dif1'].plot()
```

```python
plot_acf(df['dif1'].dropna(),lags=18, zero = False);
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.stattools import adfuller, kpss
```

```python
acorr_ljungbox(df_close, lags =[18], return_df= True)
```

```python
acorr_ljungbox(df_close, lags =18, return_df= True)
```

```python
adfuller(df_close) # is not stationary
```

```python
kpss(df_close) # is stationary
```
