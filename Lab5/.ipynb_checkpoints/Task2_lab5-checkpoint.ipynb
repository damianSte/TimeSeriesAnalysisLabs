```python
import pandas as pd
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.stattools import adfuller, kpss
```

```python
df1= pd.read_csv("kWh.csv", index_col="Month",parse_dates=True, date_format="yyyy-MMMM")
df1.sort_values(by='Month', ascending= True, inplace= True)
df1.dropna(inplace= True)
df1.head()
```

```python
df1['residential million kilowatthours'].plot()
```

```python
plot_acf(df1['residential million kilowatthours'], lags=20, zero= False);
```

```python
acorr_ljungbox(df1['residential million kilowatthours'], lags= 20, return_df=True)
```

```python
adfuller(df1)
```

```python
kpss(df1)
```

```python
df1['residential million kilowatthours_diff']= df1['residential million kilowatthours'].diff()
```

```python
plot_acf(df1['residential million kilowatthours_diff'].dropna(), zero=False);
```

```python
acorr_ljungbox(df1['residential million kilowatthours_diff'].dropna(), lags=20, return_df=True)
```

```python
acorr_ljungbox(df1['residential million kilowatthours_diff'].dropna(), lags=20, return_df=True)
```

```python
kpss(df1['residential million kilowatthours_diff'].dropna())
```
