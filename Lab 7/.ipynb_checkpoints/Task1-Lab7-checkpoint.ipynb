```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd

df = pd.read_csv("kWh_elec.csv")

df = df.iloc[::-1]

df.dropna(inplace=True)

df['Month'] = pd.to_datetime(df['Month'], format='%b %Y')
df.set_index('Month', inplace=True)
```

```python
train = df.iloc[:-12]  # Use all data except the last 12 months for training
test = df.iloc[-12:]   # Use the last 12 months for testing
```

```python
df.head()
```

```python
df['all sectors million kilowatthours'].plot()
```

```python
def plot_acf_pacf(data, lags=12):
    fig, ax = plt.subplots(2)
    plot_acf(data, lags=lags, ax=ax[0], zero=False, title='Autocorrection Function (ACF)')
    plot_pacf(data, lags=lags, ax=ax[1], zero=False, title='Partial Autocorrection Function (PACF)')
    plt.tight_layout()
    plt.show()
```

```python
plot_acf_pacf(df['all sectors million kilowatthours'])
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

df1 = df.copy()
```

```python
df1 = df.resample('MS').mean()
df_model = df1['all sectors million kilowatthours']
```

```python

sarima_model = SARIMAX(df_model, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_results = sarima_model.fit()
```

```python
hw_model = ExponentialSmoothing(train['all sectors million kilowatthours'], seasonal='additive', seasonal_periods=12)
hw_results = hw_model.fit()
```

```python
hw_predictions = hw_results.forecast(12)

# Plot the forecast for Holt-Winters model
plt.figure(figsize=(12, 6))
plt.plot(train['all sectors million kilowatthours'], label='Training data', color='blue')
plt.plot(test.index, test['all sectors million kilowatthours'], label='Test data', color='green')
plt.plot(train.index, hw_predictions, label='Fitted Training Data')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sarima_predictions = sarima_results.get_forecast(steps=12)
forecast_mean = sarima_predictions.predicted_mean
forecast_ci = sarima_predictions.conf_int()
 
# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(df_model, label='Observed')
plt.plot(forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink')

plt.title("All sectors million kilowatthours - SARIMAX model")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python

hw_model = ExponentialSmoothing(df["all sectors million kilowatthours"], trend="add", seasonal="add", seasonal_periods=12).fit()

hw_forecast = hw_model.forecast(steps=12)

hw_mape= mean_absolute_percentage_error(df['all sectors million kilowatthours'][-12:], hw_forecast)
print("Holt-Winters MAPE:", hw_mape)
```

```python
forecast = sarima_results.get_forecast(steps=12)
mean_forecast = forecast.predicted_mean
confidence_intervals = forecast.conf_int()  #uncertainty of predicament

# Calculate MAPE for SARIMA
sarima_mape= mean_absolute_percentage_error(df['all sectors million kilowatthours'][-12:], mean_forecast)
print("SARIMA MAPE:", sarima_mape)
```

```python
if hw_mape < sarima_mape:
    print("Holt-Winters model is better")
else:
    print("SARIMA model is better")
```
